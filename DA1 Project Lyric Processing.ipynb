```python
import numpy as np
import pandas as pd
import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
the_lyric_folder = 'General Lyrics'
artist_lyric_files = os.listdir(the_lyric_folder)
artist_lyric_file_paths = [os.path.join(the_lyric_folder, n) for n in artist_lyric_files]
with open(artist_lyric_file_paths[2],encoding= 'UTF-8') as f:
    song_string = f.read()
```

```python
word_cloud1 = WordCloud(collocations= False, background_color= 'white', width = 2048, height = 1080).generate(song_string)
plt.imshow(word_cloud1, interpolation='bilinear')
plt.axis("off")
plt.show()
```
